```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
my_data_filepath = "../input/covid19-healthy-diet-dataset/Food_Supply_Quantity_kg_Data.csv"

my_data= pd.read_csv(my_data_filepath, index_col='Country')
my_data.head()
```

```python
my_data.index
```

```python
my_data.columns
```

```python
#Creating a Pivot Table with Country as index and the kg of Alcoholic Bev
heatmap1_data = pd.pivot_table(my_data, 
                     index=['Country'], 
                     values='Alcoholic Beverages')
heatmap1_data
#Bar Chart #1 using the Pivot table
plt.figure(figsize=(25,30))
sns.barplot(x= heatmap1_data['Alcoholic Beverages'] ,y= heatmap1_data.index)
```

```python
#Bar chart #2 without the Pivot Table

plt.figure(figsize=(20,30))

plt.title("Alcoholic Beverages per Country in kg")

sns.barplot(y=my_data.index ,x=my_data['Alcoholic Beverages'])
```

```python
#Create a pivot table with Alcoholic bev & Obesity% 
heatmap2_data = pd.pivot_table(my_data, columns=('Country'),
                     values=['Alcoholic Beverages', 'Obesity'],
                     )
heatmap2_data
```

```python
#Creating a Scatter Plot 
plt.figure(figsize=(10,7))

sns.scatterplot(x=my_data['Obesity'], y=my_data['Animal fats'],  hue=my_data['Animal fats'])
plt.ylabel("Animal fats consuption in kg ")
plt.xlabel("Obesity % each country")
plt.title("Animal fats consumption and Obesity %")
```

```python
#With Regression Line
plt.figure(figsize=(14,6))

sns.regplot(x=my_data['Obesity'], y=my_data['Animal fats'])
plt.ylabel("Animal fats consuption in kg ")
plt.xlabel("Obesity % each country")
plt.title("Animal fats consumption and Obesity %")
```

```python
sns.regplot(x=my_data['Obesity'], y=my_data['Alcoholic Beverages'])
plt.ylabel("Alcoholic Beverages consuption in kg ")
plt.xlabel("Obesity % each country")
plt.title("Alcoholic Beverages consumption and Obesity %")
```

```python
plt.figure(figsize=(10,6))

sns.regplot(x=my_data['Obesity'], y=my_data['Vegetables'])
plt.ylabel(" Vegetables consuption in kg ")
plt.xlabel("Obesity % each country")
plt.title("Veggies consumption and Obesity %")
```
